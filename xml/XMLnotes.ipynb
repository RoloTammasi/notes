```python
from lxml import objectify
from xml.etree import ElementTree
import pandas as pd

file_name = 'baby_names_nyc.xml'

dom = ElementTree.parse(file_name)
```

```python
names = dom.findall('row/row')

for n in names:
    
    name = n.find('nm').text
    count = n.find('cnt').text
    rank = n.find('rnk').text
    print name + '\t\t' + rank
```

```python
xml = objectify.parse(open(file_name))
root = xml.getroot()
```

```python
root.getchildren()[0].getchildren()[0].getchildren()
```

```python
len(root.getchildren()[0].getchildren())
```

```python
df = pd.DataFrame(columns=('Name','Gender','Ethnicity','Count','Rank'))
```

```python
for i in range(0, len(root.getchildren()[0].getchildren())):
    obj = root.getchildren()[0].getchildren()[i].getchildren()
    row = dict(zip(['Year','Name','Gender','Ethnicity','Count','Rank'],[obj[0].text, obj[3].text, obj[1].text, obj[2].text, obj[4].text, obj[5].text]))
    row_s = pd.Series(row)
    row_s.name = i
    df = df.append(row_s)
```

```python
df.sort(['Rank','Count'], ascending=[True, False])
```

```python
from bokeh.io import output_file, show
from bokeh.models import GeoJSONDataSource
from bokeh.plotting import figure
from bokeh.sampledata.sample_geojson import geojson
import overpass
import json
```

```python
# Define pretty print function

def pp(json_thing, sort=True, indents=4):
    if type(json_thing) is str:
        print(json.dumps(json.loads(json_thing), sort_keys=sort, indent=indents))
    else:
        print(json.dumps(json_thing, sort_keys=sort, indent=indents))
    return None
```

```python
# Get data in GeoJSON format from OpenStreetMap API
#  current example = all Starbucks in Northern Virginia
#  with Internet access

api = overpass.API()
response = api.Get('node(38.70533855257685,-77.56690979003906,39.047985691121234,-77.10205078124999)["name"="Starbucks"]["internet_access"]')
```

```python
# Show JSON data

pp(response)
```

```python
# Plot data with Bokeh

source = GeoJSONDataSource(geojson=json.dumps(response))
p = figure()
p.circle(x='x', y='y', alpha=0.9, source=source)
output_file("geojson.html")
show(p)
```

```python
# Import Google Maps stuff

from bokeh.models import (
  GMapPlot, GMapOptions, ColumnDataSource, Circle, DataRange1d, PanTool, WheelZoomTool, BoxSelectTool
)
```

```python
# Plot data with Google Maps

map_options = GMapOptions(lat=38.864837, lng=-77.357876, map_type="roadmap", zoom=10)

plot = GMapPlot(
    x_range=DataRange1d(), y_range=DataRange1d(), map_options=map_options, title="NoVA"
)

source = GeoJSONDataSource(geojson=json.dumps(response))

circle = Circle(x="x", y="y", size=15, fill_color="blue", fill_alpha=0.8, line_color=None)
plot.add_glyph(source, circle)

plot.add_tools(PanTool(), WheelZoomTool(), BoxSelectTool())
output_file("gmap_plot.html")
show(plot)
```
